```python
import pandas as pd

# 데이터 불러오기
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/pyTorch/A 음식점(15,726).xlsx", engine="openpyxl")

# 개체명 중 첫번째 레이블 추출 (빈 label도 존재)
labels = [str(row).split(",")[0].strip() if not pd.isna(row) else "" for row in df["개체명"].tolist()]

# 중복 레이블 제거
unique_labels = list(set(labels)) #set 으로 중복 제거하고 다시 list를 통해 list로 만든 후 unique_labels에 저장

# 딕셔너리 형태로 만드는데 형태는 {label : idx }로 만든다
label2idx = {label : idx for idx, label in enumerate(unique_labels)}

label2idx
```

```python
# 중복되지 않는 레이블을 추출하여 인덱스 매핑 딕셔너리 생성
unique_labels = list(set(labels))
label2idx = {label: idx for idx, label in enumerate(unique_labels)}

# 레이블을 인덱스로 변환
label_indices = [label2idx[label] for label in labels]

# 결과 확인
print((label_indices))
```

```python
# label2idx 딕셔너리를 저장하고 나중에 사용할 수 있도록 저장
import json

with open("label2idx.json", "w") as json_file:
    json.dump(label2idx, json_file)

# 나중에 label2idx 불러오기
with open("label2idx.json", "r") as json_file:
    loaded_label2idx = json.load(json_file)
```
