```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
import json

# 데이터프레임을 읽어옵니다.
excel_file = pd.read_excel("C:\\Users\\gjaischool\\Desktop\\A 음식점(15,726).xlsx", engine="openpyxl")


# PyTorch 사용하여 커스텀 데이터셋 만들기 
# 데이터셋 클래스를 정의합니다.
class CustomDataset(Dataset): # Dataset 클래스 상속
    def __init__(self, data, transform=None, target_transform=None): # init 메서드 구현
        self.df = data  # 입력받은 data를 클래스 내 변수 self.df에 저장
        self.sentence, self.label = self.whatisthis(self.df) # whatisthis 메서드에서 가지고와서
        # 클래스 내 변수 sentencfe 와 label에 각각 저장

    def whatisthis(self, excel_df):
        sentence = excel_df["SENTENCE"] # 우리 excel파일의 sentence에서 sentence 저장
        label = [str(row).split(",")[0].strip() if not pd.isna(row) else "" for row in excel_df["개체명"].tolist()]
        # excel파일의 개체명 중 첫번째를 label에 저장하고 빈 공백도 빈 공백대로 저장

        return sentence, label

    def __len__(self): # len 메서드 구현
        return len(self.df) # len함수로 데이터셋의 전체 길이 반환 ( 데이터셋의 크기를 반환)

    def __getitem__(self, idx): # getitem 메서드 구현
        sentence = self.sentence[idx] # 데이터셋 특정 sentence인덱스를 sentence에 넣고
        label = self.label[idx] # 데이터셋 특정 label인덱스를 label에 넣고

        return sentence, label # 그 sentence 와 label을 반환

# 데이터셋 인스턴스 생성
custom_dataset = CustomDataset(excel_file) # 커스텀한 데이터를 로드하기 위해 데이터셋 인스턴스 생성

# 데이터 로더를 생성
batch_size = 64  # 원하는 배치 크기로 설정
data_loader = DataLoader(custom_dataset, batch_size=batch_size, shuffle=True)  
# DataLoader 클래스 사용  앞서 만든 custom_dataset 객체를 기반으로 미니배치를 생성
# batch_size를 활용해 미니배치에 포함되는 데이터의 개수 지정
# shuffle을 통해 무작위 True 설정

# 모든 데이터를 저장할 리스트를 만듭니다.
data_set = []

# 데이터 로더를 사용하여 데이터를 반복(iterate)하면서 모델을 학습할 수 있습니다.
for batch in data_loader:
    sentences, labels = batch
    # 각 sentence와 label을 짝지어서 리스트에 추가
    for sentence, label in zip(sentences, labels): # zip() : 2개 이상의 객체를 묶어주는 역할 
        data_set.append({"sentence": sentence, "label": label})
        print("Sentence:", sentence) # 각 문장 출력
        print("Label:", label) # 각 라벨 출력

# JSON 파일로 저장 (ensure_ascii=False를 추가하여 한글 문자열을 유니코드가 아닌 형식으로 저장)
with open("data.json", "w", encoding='utf-8') as f:
    json.dump(data_set, f, ensure_ascii=False, indent=4)
```
